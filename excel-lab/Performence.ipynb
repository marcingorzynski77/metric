```python
def make_categorical(df, col_name):
    df.loc[:, col_name] = pd.Categorical(df[col_name]) 
```

```python
def mem_mib(df):
    print("{0:.2f} MiB".format(
        df.memory_usage().sum() / (1024 * 1024)
    ))
```

```python
make_categorical(df1_quick, "PartyName")
make_categorical(df1_quick, "InstrumentTenor")
make_categorical(df1_quick, "RiskMeasureType")
make_categorical(df1_quick, "UnderlyingTenor")

mem_mib(df1_quick)
```
